```python
# data cleaning for dataset in regression model
#importing module
import pandas as pd
import matplotlib
import matplotlib.pyplot as pp

df = pd.read_csv('d:/healthcare-dataset-stroke-data.csv')
df
df.head()
```

```python
df.columns
#displayinf last five rows of dataset
df.tail()
```

```python
#Rebuild Missing Data
df.isnull()
```

```python
#finding null value column wise 
df.isna().any()
#we can infer that null values are present only in bmi column
```

```python
# finding the count for null values
df.isna().sum()
#totally 201 null values are there in bmi
```

```python
# check duplicate values
df.duplicated()
```

```python
#droppping null value column 
df2 = df.dropna(axis='columns') 
print(df2)
#the column dropped is bmi
```
